# **Ejercicio 1**

El código de barras es un código basado en la representación de un conjunto de líneas paralelas de distinto grosor, color y espaciado que en su conjunto contienen una determinada información. De este modo, el código de barras permite reconocer rápidamente un artículo de forma única, global y no ambigua en un punto de la cadena logística y así poder realizar inventario o consultar sus características asociadas.

Existen principalmente dos formatos, que se diferencian en el ancho. El formato EAN-8 codifica 8 números, y el formato EAN-13, codifica 13.

En cada caso, el último dígito (el de más a la derecha) se usa como dígito de verificación y se calcula con base en los demás.

* Empezando por la derecha (omitiendo el dígito de verificación que se está calculando), se suman los dígitos individuales, multiplicados por un factor.

>* Los dígitos en posiciones impares (empezando por la derecha y omitiendo el dígito verificación) se multiplican por 3.
>* Los dígitos en posiciones pares se multiplican por 1.

* Por ejemplo, para el código EAN-8 de la figura la operación a realizar es la siguiente.

$(1 * 3) + (0 * 1) + (1 * 3) + (1 * 1) + (3 * 3) + (0 * 1) + (9 * 3) = 43
$


El dígito de verificación corresponde al número que hay que sumar al resultado anterior para que sea un valor múltiplo de 10. En este caso, para llegar al múltiplo de 10 más cercano mayor a 43, hay que sumar 7 para llegar a 50. En caso de que la suma de como resultado un múltiplo de 10, el dígito de verificación es 0 (cero).
